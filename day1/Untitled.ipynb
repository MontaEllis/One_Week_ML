```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
data_all = pd.read_csv("data_all.csv")
data_all.head()
```

```python
features = [x for x in data_all.columns if x not in ['status']]
features
```

```python
x = data_all[features]
y = data_all['status']
x
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(x,y,test_size=0.3,random_state = 2018)
```

```python
X_train
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression(random_state =2018)
lr.fit(X_train,Y_train)
lr_result = lr.score(X_test,Y_test)
lr_result
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
tr = DecisionTreeClassifier(random_state=2018)
tr.fit(X_train,Y_train)
tr_result = tr.score(X_test,Y_test)
tr_result
```

```python
from sklearn.svm import LinearSVC
```

```python
svm = LinearSVC(random_state=2018)
svm.fit(X_train,Y_train)
svm_result = svm.score(X_test,Y_test)
svm_result
```
