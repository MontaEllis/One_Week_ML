```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,roc_auc_score,roc_curve

```

```python
import matplotlib.pyplot as plt
```

```python
file = pd.read_csv("data_all.csv")
file.info()
```

```python
from sklearn.preprocessing import StandardScaler
features= [x for x in file.columns if x not in ['status']]
x = file[features]
y = file['status']
X_train, X_test, Y_train, Y_test = train_test_split(x,y,test_size=0.3,random_state = 2018)

# 归一化
scaler = StandardScaler()
scaler.fit(X_train)
X_train_standard = scaler.transform(X_train)
X_test_standard = scaler.transform(X_test)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
def grid_search(model,parameters): 
    grid=GridSearchCV(model,parameters,scoring='accuracy',cv=5) 
    grid=grid.fit(X_train_standard,Y_train) 
    if hasattr(model,'decision_function'): 
        pre_test=grid.decision_function(X_test_standard) 
    else: 
        pre_test=grid.predict_proba(X_test_standard)[:,1] 
    print('best score: %f',grid.best_score_) 
    print(grid.best_params_) 
    print('test score: %f',grid.score(X_test_standard, Y_test)) 
    print("AUC",roc_auc_score(Y_test,pre_test))
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
parameter = {'C':[0.1,1,2,3],'penalty':['l1','l2']}
lr = LogisticRegression(random_state=2018)
grid_search(lr, parameters=parameter)
```

```python
from sklearn.svm import LinearSVC
```

```python
parameter = {'tol':[1e-3,1e-4],'C':[0.1,1,5]}
svm = LinearSVC(random_state=2018)
grid_search(svm,parameters=parameter)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
tree=DecisionTreeClassifier(random_state=2018)
parameter={'max_depth':range(1,100),'criterion':['gini','entropy'],'max_features':['sqrt','log2','auto']}
grid_search(tree,parameters=parameter)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(random_state=2018)
parameter = {'max_depth': range(1, 5)}
grid_search(rf,parameters=parameter)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
gbdt=GradientBoostingClassifier(random_state=2018)
parameter={'n_estimators':range(30,80,10),'max_depth':range(3,10)}
grid_search(gbdt,parameters=parameter)
```

```python
from xgboost.sklearn import XGBClassifier
import numpy as np
```

```python
xgbc = XGBClassifier(random_state=2018)
parameter={'eta':np.arange(0.1,0.4,0.1),'min_child_weight':range(1,6,2),'gamma':[0.1,0.2]}
grid_search(xgbc,parameters=parameter)
```

```python
import lightgbm as lgb
```

```python
lgb_model = lgb.LGBMClassifier(boosting_type='GBDT',random_state=2018,silent=0)
parameter={'max_depth':range(3,10),'n_estimators=10':[25,30,35]}
grid_search(lgb_model,parameters=parameter)
```
